```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt; plt.rcParams["figure.figsize"] = (30,15)
```

```python
biometrics = pd.read_csv('../data/biometrics/biometrics.csv')

X = [['height', 'wingspan'], ['height', 'weight'],
     ['age', 'bmi'], 
     ['weight', 'waist_circumference'],
     ['height', 'torso_length'],
     ['shoulder_circumference', 'waist_circumference'],
     ['height', 'shoe_size'], 
     ['femur_length', 'leg_length']
     ]

fig, axes = plt.subplots(len(X), 1, figsize = (30, len(X)*15))
for i, x in enumerate(X):    
    sns.regplot(x = x[0], y = x[1],order = 2, data = biometrics, ax=axes[i]);
    axes[i].set_title(x[0] + ' vs ' + x[1], fontsize = 24);
```
